```python
from PIL import Image, UnidentifiedImageError
import os
```

```python
# Input folder containing the images
input_folder = "../data/final-data/images/4noktaWinter47-48-2k"

# Output folder for the converted images
output_folder = "../data/final-data/images/png/4noktaWinter47-48-2k"
```

```python
# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)
```

```python
# Initialize counters
input_image_count = 0
output_image_count = 0
```

```python
# Iterate through all files in the input folder
for filename in os.listdir(input_folder):
    if filename.endswith(".tiff") or filename.endswith(".tif"):
        # Increment the input image count
        input_image_count += 1

        try:
            # Load the TIFF image
            tiff_filepath = os.path.join(input_folder, filename)
            img = Image.open(tiff_filepath)

            # Convert the image to PNG format
            png_filename = os.path.splitext(filename)[0].lower() + ".png"
            png_filepath = os.path.join(output_folder, png_filename)
            img.save(png_filepath, "PNG")

            # Increment the output image count
            output_image_count += 1

            print(f"Converted: {filename} -> {png_filename}")

        except UnidentifiedImageError as e:
            print(f"Error processing {filename}: {e}")

print("Conversion and renaming completed.")
print(f"Input Image Count: {input_image_count}")
print(f"Output Image Count: {output_image_count}")
```
