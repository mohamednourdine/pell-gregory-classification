```python
import os
import csv
import pandas as pd
```

```python
# Define the input dataset file and output directory
input_csv_file_3738 = "../data/final-data/annotations/37-38-WINTER-KODAK.csv"

# Define the input dataset file and output directory
input_csv_file_4748 = "../data/final-data/annotations/47-48-WINTER-KODAK.csv"

output_directory = "../data/final-data/annotations/37-38-WINTER-KODAK"
```

```python
# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)
```

```python
def convert_image_names_to_lowercase(csv_file_path):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)
     print(df.columns)
    # Check if the "IMAGE" column exists
    if 'IMAGE' in df.columns:
        # Convert the names in the "IMAGE" column to lowercase
       
#         df['IMAGE'] = df['IMAGE'].str.lower()

        # Save the changes back to the CSV file
#         df.to_csv(csv_file_path, index=False)
#         print(f'Successfully converted and saved {csv_file_path}')
    else:
        print('The "IMAGE" column does not exist in the CSV file.')
```

```python
# Function to save annotations to a file
def save_annotations(image_name, annotations):
    annotation_text = "\n".join(annotations)
    with open(os.path.join(output_directory, f"{image_name}.txt"), "w") as output_file:
        output_file.write(annotation_text)
```

```python
os.makedirs(output_directory, exist_ok=True)
```

```python
# Read the CSV file and process each row
with open(input_csv_file_3738, mode="r", newline="", encoding="utf-8-sig") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    for row in csv_reader:
        # Get the image name from the "IMAGE" column
        image_name = row.get("IMAGE")

        # Skip rows with missing or empty image names
        if not image_name:
            print(f"Skipping row with missing image name: {row}")
            continue

        # Create a list to store the coordinates
        coordinates = []

        # Iterate through the columns (1-X, 1-Y, 2-X, 2-Y, etc.) and extract the coordinates
        for col in range(1, 6, 1):
            x_col_name = f"{col}-X"
            y_col_name = f"{col}-Y"
            print(x_col_name,y_col_name)


            x = row.get(x_col_name)
            y = row.get(y_col_name)

            # Skip rows with missing or empty coordinates
            if not x or not y:
                print(f"Skipping row with missing coordinates: {row}")
                continue

            coordinates.append(f"{x},{y}")

        # Create a TXT file for the image and write the coordinates
        if coordinates:
            output_file_path = os.path.join(output_directory, f"{image_name}.txt")
            with open(output_file_path, "w") as output_file:
                output_file.write("\n".join(coordinates))
                
print("Annotations saved successfully.")
```
