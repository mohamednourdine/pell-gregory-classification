```python
import os
from PIL import Image
import shutil
import csv
```

```python
# Path to folder tif
folder_A = "../data/dataset/images/tif/4noktaWinter37-38-2k"
folder_B = "../data/dataset/images/tif/4noktaWinter47-48-2k"
folder_C = "../data/dataset/images/tif/5noktaPellgregory37-38-2k"
folder_D = "../data/dataset/images/tif/5noktaPellgregory47-48-2k"

image_A = "../data/dataset/images/jpg"
```

```python
def convert_image_names_to_lowercase(folder_path):
    # Iterate through files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file is an image (you may add more extensions if needed)
        if filename.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tif', '.tiff')):
            # Construct the old and new file paths
            old_path = os.path.join(folder_path, filename)
            new_filename = filename.lower()
            new_path = os.path.join(folder_path, new_filename)
            
            # Rename the file
            os.rename(old_path, new_path)
            print(f"Renamed '{filename}' to '{new_filename}'")
```

```python
convert_image_names_to_lowercase(image_A)
```

```python
def resize_images(folder_A, folder_B):
    # Get list of image files in folder A
    images_A = os.listdir(folder_A)
    
    # Iterate through images in folder A
    for img_name in images_A:
        img_path_A = os.path.join(folder_A, img_name)
        img_path_B = os.path.join(folder_B, img_name.replace(".tif", ".jpg"))  # Assuming names are same except for extension
        
        # Open image from folder A
        img_A = Image.open(img_path_A)
        
        try:
            # Open corresponding image from folder B
            img_B = Image.open(img_path_B)
            
            # Check if image B has the same size as image A
            if img_B.size == img_A.size:
                # print(f"Skipping '{img_name}': Images have the same size")
                continue
            
            # Resize image B to match size of image A
            img_B_resized = img_B.resize(img_A.size, resample=Image.BILINEAR)
            
            # Save resized image B
            img_B_resized.save(img_path_B)
            
            # Print details about resizing operation
            print(f"Resized '{img_name}': Original size {img_B.size}, New size {img_A.size}")
        
        except FileNotFoundError:
            print(f"Image not found: {img_path_B}")
```

```python
resize_images(folder_A, image_A)
```

```python
resize_images(folder_B, image_A)
```

```python
resize_images(folder_C, image_A)
```

```python
resize_images(folder_D, image_A)
```

```python
def lowercase_no_column(csv_path):
    # Read the CSV file
    with open(csv_path, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        rows = list(reader)
    
    # Lowercase the values in the "NO" column
    for row in rows:
        row['NO'] = row['NO'].lower()
    
    # Write the updated rows back to the CSV file
    with open(csv_path, 'w', newline='') as csvfile:
        fieldnames = reader.fieldnames
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(rows)
```
