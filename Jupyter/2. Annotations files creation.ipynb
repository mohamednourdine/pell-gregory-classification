```python
# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)
```

```python
def convert_image_names_to_lowercase(csv_file_path):
    # Load the CSV file into a DataFrame
    df = pd.read_csv(csv_file_path)
    print(df.columns)
    # Check if the "IMAGE" column exists
    if 'IMAGE' in df.columns:
        # Convert the names in the "IMAGE" column to lowercase
       
        df['IMAGE'] = df['IMAGE'].str.lower()

        # Save the changes back to the CSV file
        df.to_csv(csv_file_path, index=False)
        print(f'Successfully converted and saved {csv_file_path}')
    else:
        print('The "IMAGE" column does not exist in the CSV file.')
```

```python
# Function to save annotations to a file
def save_annotations(image_name, annotations):
    annotation_text = "\n".join(annotations)
    with open(os.path.join(output_directory, f"{image_name}.txt"), "w") as output_file:
        output_file.write(annotation_text)
```

```python
# Define the input dataset file and output directory
one = "../data/dataset/annotations/37-38-WINTER.csv"
one_output = "../data/dataset/annotations/37-38-WINTER"
os.makedirs(one_output, exist_ok=True)

# Define the input dataset file and output directory
two = "../data/dataset/annotations/47-48-WINTER.csv"
two_output = "../data/dataset/annotations/47-48-WINTER"
os.makedirs(two_output, exist_ok=True)

three = "../data/dataset/annotations/37-38-IlkGoruntuler-PG.csv"
three_output = "../data/dataset/annotations/37-38-IlkGoruntuler-PG"
os.makedirs(three_output, exist_ok=True)

four = "../data/dataset/annotations/47-48-IlkGoruntuler-PG.csv"
four_output = "../data/dataset/annotations/47-48-IlkGoruntuler-PG"
os.makedirs(four_output, exist_ok=True)

five = "../data/dataset/annotations/37-38-IlkGoruntuler-Winter.csv"
five_output = "../data/dataset/annotations/37-38-IlkGoruntuler-Winter"
os.makedirs(five_output, exist_ok=True)

six = "../data/dataset/annotations/47-48-IlkGoruntuler-Winter.csv"
six_output = "../data/dataset/annotations/47-48-IlkGoruntuler-Winter"
os.makedirs(six_output, exist_ok=True)

# Define the input dataset file and output directory
seven = "../data/dataset/annotations/37-38-PELLGREGORY.csv"
seven_output = "../data/dataset/annotations/37-38-PELLGREGORY"
os.makedirs(seven_output, exist_ok=True)

# Define the input dataset file and output directory
eight = "../data/dataset/annotations/47-48-PELLGREGORY.csv"
eight_output = "../data/dataset/annotations/47-48-PELLGREGORY"
os.makedirs(eight_output, exist_ok=True)
```

```python
# Read the CSV file and process each row
with open(eight, mode="r", newline="", encoding="utf-8-sig") as csv_file:
    csv_reader = csv.DictReader(csv_file)
    
    for row in csv_reader:
        # Get the image name from the "IMAGE" column
        image_name = row.get("IMAGE")

        # Skip rows with missing or empty image names
        if not image_name:
            print(f"Skipping row with missing image name: {row}")
            continue

        # Create a list to store the coordinates
        coordinates = []

        # Iterate through the columns (1-X, 1-Y, 2-X, 2-Y, etc.) and extract the coordinates
        for col in range(1, 6, 1):
            x_col_name = f"{col}-X"
            y_col_name = f"{col}-Y"
            print(x_col_name,y_col_name)


            x = row.get(x_col_name)
            y = row.get(y_col_name)

            # Skip rows with missing or empty coordinates
            if not x or not y:
                print(f"Skipping row with missing coordinates: {row}")
                continue

            coordinates.append(f"{x},{y}")

        # Create a TXT file for the image and write the coordinates
        if coordinates:
            output_file_path = os.path.join(eight_output, f"{image_name}.txt")
            with open(output_file_path, "w") as output_file:
                output_file.write("\n".join(coordinates))
                
print("Annotations saved successfully.")
```
