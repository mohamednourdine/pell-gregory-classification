```python
import os
import shutil
```

```python
def list_files(folder_path):
    # Check if the folder path exists
    if not os.path.isdir(folder_path):
        return "Folder path does not exist"

    # Get list of files in the folder
    files_list = os.listdir(folder_path)

    # Replace ".txt" with ".png" in file names
    files_list = [file.replace('.txt', '.png') for file in files_list]

    # Join the file names into a comma-separated string
    files_str = ', '.join(files_list)

    return files_str
```

```python
def copy_images(source_folder, destination_folder, file_names):
    # Check if the source folder path exists
    if not os.path.isdir(source_folder):
        return "Source folder path does not exist"

    # Create the destination folder if it does not exist
    if not os.path.isdir(destination_folder):
        os.makedirs(destination_folder)
        print(f"Destination folder created: {destination_folder}")

    # Iterate through the list of file names
    for file_name in file_names:
        # Construct the source file path
        source_file_path = os.path.join(source_folder, file_name)

        # Check if the source file exists
        if not os.path.isfile(source_file_path):
            print(f"File {file_name} does not exist in the source folder.")
            continue

        # Construct the destination file path
        destination_file_path = os.path.join(destination_folder, file_name)

        # Copy the file from source to destination
        shutil.copyfile(source_file_path, destination_file_path)
        print(f"File {file_name} copied successfully.")

    return "Image copy process completed."
```

```python
folder_path = '../data/dataset/resized/annotations/37-38-PELLGREGORY'
files_str = list_files(folder_path)
file_names = files_str.split(', ')

source_folder = '../data/dataset/resized/256x256'
destination_folder = '../data/dataset/resized/37-38-PELLGREGORY'

copy_images(source_folder, destination_folder, file_names)
```
